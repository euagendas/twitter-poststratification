# NUTS3 - Merge census and Twitter with joint distributions

## Census data (joint distributions)

```python
dfCensusJoint=pd.read_csv("data/nuts3_matched_to_m3_agebrackets.csv")
dfCensusJoint.head()
```

```python
dfCensusJoint=dfCensusJoint.rename(index=str,columns={"GEO":"nuts3","VALUE":"census"})
dfCensusJoint.head()
dfCensusJoint
```

```python
dfCensusJoint["age_0017"]=dfCensusJoint["AGE_M3"].apply(lambda x: 1 if x=="<18" else 0)
dfCensusJoint["age_1829"]=dfCensusJoint["AGE_M3"].apply(lambda x: 1 if x=="18-29" else 0)
dfCensusJoint["age_3039"]=dfCensusJoint["AGE_M3"].apply(lambda x: 1 if x=="30-39" else 0)
dfCensusJoint["age_4099"]=dfCensusJoint["AGE_M3"].apply(lambda x: 1 if x=="40+" else 0)
dfCensusJoint["gender_F"]=dfCensusJoint["SEX"].apply(lambda x: 1 if x=="F" else 0)
dfCensusJoint["gender_M"]=dfCensusJoint["SEX"].apply(lambda x: 1 if x=="M" else 0)
dfCensusJoint.head()
```

```python
dfCensusJoint=dfCensusJoint.drop(["SEX","AGE_M3"],axis=1)
dfCensusJoint.head()
```

## Twitter data

```python
#dfTwLong=pd.wide_to_long(dfTw,stubnames="tw_",i=["nuts3","organizations"],j="twitter2")
tw_cols=[x[0:3]=="tw_" and not "unk" in x for x in dfTw.columns]
tw_cols=dfTw.columns[tw_cols]
dfNoZero=dfTw.iloc[np.all(dfTw[tw_cols].values!=0,axis=1),]
dfTwLong=pd.melt(dfNoZero,id_vars=["nuts3","organizations"],value_vars=tw_cols,value_name="twitter")
dfTwLong.head()
```

```python
dfTwLong["age_0017"]=dfTwLong["variable"].apply(lambda x: 1 if x[0:7]=="tw_0017" else 0)
dfTwLong["age_1829"]=dfTwLong["variable"].apply(lambda x: 1 if x[0:7]=="tw_1829" else 0)
dfTwLong["age_3039"]=dfTwLong["variable"].apply(lambda x: 1 if x[0:7]=="tw_3039" else 0)
dfTwLong["age_4099"]=dfTwLong["variable"].apply(lambda x: 1 if x[0:7]=="tw_4099" else 0)
dfTwLong["gender_F"]=dfTwLong["variable"].apply(lambda x: 1 if x[-1]=="F" else 0)
dfTwLong["gender_M"]=dfTwLong["variable"].apply(lambda x: 1 if x[-1]=="M" else 0)
dfTwLong.head()
```

```python
dfTwLong=dfTwLong.drop("variable",axis=1)
dfTwLong.head()
```

```python
dfTwCensusLong=dfTwLong.merge(dfCensusJoint,on=["nuts3","age_0017","age_1829","age_3039","age_4099","gender_F","gender_M"])
dfTwCensusLong.head()
```

```python
dfTwCensusLong["twitter"]=np.log(dfTwCensusLong["twitter"])
dfTwCensusLong["census"]=np.log(dfTwCensusLong["census"])
dfTwCensusLong["diff_census_twitter"]=dfTwCensusLong["census"]-dfTwCensusLong["twitter"]
dfTwCensusLong.head()
```

```python
#Some census values are NaN. Drop
dfTwCensusLong=dfTwCensusLong.dropna()
```

```python
dfTwCensusLong.to_csv("data/nuts3_df_data_NiXZ.orgs-as-humans.csv",index=False)
```

## Prepare dataset for debiasing in R

```python
df_data_Ni = pd.read_csv("data/nuts3_df_data_Ni.orgs-as-humans.csv")
df_data_NiXZ = pd.read_csv("data/nuts3_df_data_NiXZ.orgs-as-humans.csv")

# get age/gender marginal counts 
df_data_Ni['age_0017'] = df_data_Ni['tw_0017M'] + df_data_Ni['tw_0017F'] + df_data_Ni['tw_0017unk']
df_data_Ni['age_1829'] = df_data_Ni['tw_1829M'] + df_data_Ni['tw_1829F'] + df_data_Ni['tw_1829unk']
df_data_Ni['age_3039'] = df_data_Ni['tw_3039M'] + df_data_Ni['tw_3039F'] + df_data_Ni['tw_3039unk']
df_data_Ni['age_4099'] = df_data_Ni['tw_4099M'] + df_data_Ni['tw_4099F'] + df_data_Ni['tw_4099unk']
df_data_Ni['gender_F'] = df_data_Ni['tw_0017F'] + df_data_Ni['tw_1829F'] + df_data_Ni['tw_3039F'] + df_data_Ni['tw_4099F'] +df_data_Ni['tw_unkF'] 
df_data_Ni['gender_M'] = df_data_Ni['tw_0017M'] + df_data_Ni['tw_1829M'] + df_data_Ni['tw_3039M'] + df_data_Ni['tw_4099M'] +df_data_Ni['tw_unkM']
df_data_Ni.head()
```

```python
# add country code field to df_data_Ni
nuts3_cities = df_data_Ni['nuts3'].values
country =  [city[:2] for city in nuts3_cities]
df_data_Ni['country'] = country
print('# countries: ', len(set(country)))

# Dropping data from countries with only one data sample or without one joint demographic (age/gender) count
for country_code in ['CY','LI','SI']:
    df_data_Ni = df_data_Ni.drop(df_data_Ni[df_data_Ni.country==country_code].index)
   
print('# countries: ', len(set(df_data_Ni['country'].values)))
print(set(df_data_Ni['country'].values))


# add country code field to df_data_NiXZ
nuts3_cities = df_data_NiXZ['nuts3'].values
country =  [city[:2] for city in nuts3_cities]
df_data_NiXZ['country'] = country
print('# countries: ', len(set(country)))

for country_code in ['CY','LI','SI']:
    df_data_NiXZ = df_data_NiXZ.drop(df_data_NiXZ[df_data_NiXZ.country==country_code].index)

print('# countries: ', len(set(df_data_NiXZ['country'].values)))
df_data_NiXZ = df_data_NiXZ.sort_values('nuts3')
```

### Aggregate counts for different debiasing models

```python
# use the same number of nut3-regions for the two dataframes (df_data_Ni and df_data_NiXZ)
Ni_regions = df_data_Ni['nuts3'].unique()
NiXZ_regions = df_data_NiXZ['nuts3'].unique()
common_regions  = list(set(Ni_regions) & set(NiXZ_regions))

df_data_Ni = df_data_Ni[df_data_Ni['nuts3'].isin(common_regions)]
df_data_NiXZ = df_data_NiXZ[df_data_NiXZ['nuts3'].isin(common_regions)]

df_data_Ni.sort_values(by=['nuts3'])
df_data_NiXZ.sort_values(by=['nuts3'])
```

```python
df_data_NiXZ['age'] = df_data_NiXZ[['age_0017','age_1829','age_3039','age_4099']].idxmax(axis=1)
df_data_NiXZ['gender'] = df_data_NiXZ[['gender_F','gender_M']].idxmax(axis=1)
df_data_NiXZ.head()
```

```python
df_data_Ni.to_csv("data/P_nuts3_df_data_Ni.orgs-as-humans.csv")
df_data_NiXZ.to_csv("data/P_nuts3_df_data_NiXZ.orgs-as-humans.csv")
```
